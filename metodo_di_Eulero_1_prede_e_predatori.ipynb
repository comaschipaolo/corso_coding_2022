# Problemi di popolazioni

Nel 1925 il biologo italiano Umberto D'Ancona studi l'andamento delle popolazioni ittiche nell'alto Adriatico e noto uno strano fenomeno: la riduzione della pesca dovuta alla prima guerra mondiale (1915-18) ha causato un aumento relativo delle popolazioni di predatori (come gli squali) a svantaggio dei pesci preda. Questo fenomeo  controintuitivo, dato che l'attivit di pesca colpisce soprattutto le prede e non i predatori.



La scienza si propone di investigare fenomeni di questo tipo in modo quantitativo, utilizzando numeri, formule e calcoli. La traduzione di un fenomeno in concetti matematici  un *modello matematico*. La parte della matematica che si occupa di studiare i modelli matematici  la matematica applicata.

Generalmente un modello matematico semplifica la complessit intrinseca al fenomeno, rinunciando a descriverne in modo completo levoluzione, per provare a coglierne alcuni aspetti. Un matematico applicato cercher sempre di mantenere un equilibrio tra la complessit del modello e laffidabilit dei risultati ottenuti, anche in base alle risorse computazionali a disposizione (da carta e penna fino ai supercomputer).




La costruzione di un modello passa, quindi, per varie fasi:

* **semplificazione**: in questa fase si cerca un compromesso accettabile tra complessit del modello e aderenza alla realt. E' importante capire quali sono gli aspetti che pi ci interessano e perch possiamo trascurarne altri. Ad esempio, in questa fase dovremo decidere quante specie considerare, se includere o meno la stagionalit del clima o l'interazione con l'uomo, etc...

* **modellizzazione**: fissate le variabili da tenere in considerazione, si scrivono le equazioni matematiche che regolano le relazioni tra tali variabili e la loro evoluzione nel tempo.

* **risoluzione**: la risoluzione delle equazioni pu essere di tipo algebrico o numerico (mediante l'utilizzo di un calcolatore). Pu essere esatta o approssimata.

* **validazione**: il modello, mediante la risoluzione, ci permette di calcolare dei risultati quantitativi. Questi risultato devono essere confrontati con i dati reali, per validare il modello e/o cercare di migliorarlo.

## Modello I: la dinamica naturale di una popolazione

Iniziamo studiando il modello di popolazione pi semplice possibile: consideriamo un'unica specie e trascuriamo l'influenza dell'ambiente sulla sua evoluzione. Quindi, le risorse (cibo, spazio vitale, etc...) a disposizione di questa specie sono illimitate. Inoltre, consideriamo l'ambiente in cui vive la specie come chiuso (non ci possono essere scambi di elementi della specie con l'esterno).

Introduciamo i seguenti concetti e le seguenti definizioni, che ci porteranno a modellizzare il problema. Nel seguito chiamiamo $y(t)$ il numero di elementi della popolazione in funzione del tempo.

* **Definizione:** il *tasso specifico di natalit* $n$  il rapporto tra il numero di nuovi nati in una unit di tempo e il numero medio di individui della popolazione nello stesso periodo.

Se consideriamo un intervallo di tempo $\Delta t$, il numero medio di nati nell'intervallo di tempo  

$$\Delta y_n(t) \approx n\cdot y(t)\cdot \Delta t$$.

Questa relazione  approssimata, perch la popolazione $y(t)$ varia nell'intervallo di tempo $\Delta t$.

* **Definizione:** il *tasso specifico di mortalit* $m$  il rapporto tra il numero di morti naturali in una unit di tempo e il numero medio di individui della popolazione nello stesso periodo.

Analogamente, il numero medio di morti nell'intervallo di tempo  

$$\Delta y_m(t) \approx m\cdot y(t)\cdot \Delta t$$.


* **Definizione:** la differenza
$$r = n-m$$
 il *tasso di crescita netto per unit di popolazione*.

La variazione complessiva della popolazione nell'intervallo di tempo $\Delta t$  data dalla differenza tra nati e morti nell'intervallo di tempo:

$$\Delta y(t) \approx \Delta y_n(t) - \Delta y_m(t) = (n-m)y(t)\cdot \Delta t = r\cdot y(t) \cdot \Delta t$$

e quindi

$$\frac{\Delta y}{\Delta t} \approx r\cdot y(t)$$

**Per chi  in quinta:** ovviamente, nel limite per $\Delta t \rightarrow 0$ il rapporto nel primo membro dell'equazione precedente  la derivata di $y(t)$ e la relazione diventa esatta

$$\frac{dy}{dt} = y'(t) = r\cdot y(t)$$

L'equazione che abbiamo trovato governa la variazione della popolazione in funzione del tempo. Per poter calcolre effettivamente $y(t)$  necessario introdurre un'altra infomazione: la popolazione all'istante iniziale.

### Legge di Malthus

Questo sistema di equazioni  detto **legge di Malthus**:

$$\begin{cases}
  \frac{\Delta y}{\Delta t} \approx r\cdot y(t) \\
  y(0) = y_0
\end{cases}$$

Un sistema di questo tipo  detto **problema di Cauchy** ed  un esempio di **equazione differenziale**. Un'equazione differenziale  un problema matematico estremamente complesso perch l'incognita non  un numero, ma  una funzione (in questo caso, la funzione $y(t)$). Tutti i modelli nelle scienze quantitative si basano sulla formulazione e la risoluzione di equazioni differenziali, in particolare la Fisica e l'Economia.

In questo caso particolare, esiste una soluzione esatta al sistema:

$$y(t) = y_0\cdot e^{rt}$$

Il seguente codice rappresenta in un grafico la soluzione della legge di Malthus con $y_0 = 100$ e nei casi $r_1 = 0,1$, $r_2 = 0,01$ e $r_3 = -0,1$.

```python
from matplotlib import pyplot as plt
import numpy as np
import math



t_arr = np.linspace(0, 20, 40)

y_0 = 100
r_1 = 0.1
r_2 = 0.01
r_3 = -0.1

y_1 = [y_0*np.e**(r_1*t) for t in t_arr]
y_2 = [y_0*np.e**(r_2*t) for t in t_arr]
y_3 = [y_0*np.e**(r_3*t) for t in t_arr]


# disegna le linee dei tre grafici
plt.plot(t_arr, y_1, label="r = 0.1")
plt.plot(t_arr, y_2, label="r = 0.01")
plt.plot(t_arr, y_3, label="r = -0.1")

# etichette agli assi
plt.xlabel("tempo (anni)")
plt.ylabel("popolazione")

# titolo e legenda
plt.title("La legge di Malthus")
plt.legend()

# attiva una griglia
plt.grid()

plt.show()
```

# Il metodo di Eulero
La legge di Malthus  un esempio di equazione differenziale che si pu risolvere in modo esatto. Tuttavia, solo una piccolissima parte delle equazioni differenziali possono essere risolte in modo esatto: in generale i problemi scientifici richiedono equazioni differenziali che non hanno soluzioni esatte, ma le cui soluzioni possono essere approssimate dal computer con metodi numerici.

Il metodo di Eulero  unapprossimazione che permette di integrare numericamente equazioni differenziali di questo tipo. Esso si basa sullidea di discretizzare il tempo, cio di suddividere il tempo in intervalli identici di ampiezza $\Delta t$:

$$\begin{cases}
  t_0 = 0 \\
  t_1 = t_0 + \Delta t = \Delta t\\
  t_2 = t_1 + \Delta t = 2\Delta t\\
  t_3 = t_2 + \Delta t = 3\Delta t\\
  ...\\
  t_N = t_{N-1} + \Delta t = N\Delta t
\end{cases}$$

In questo modo, se consideriamo la legge di Malthus o un problema di Cauchy pi generale

$$\begin{cases}
  \frac{\Delta y}{\Delta t} \approx f(t,y(t)) \\
  y(0) = y_0
\end{cases}$$

Possiamo discretizzare il rapporto al primo membro 

$$\frac{\Delta y(t_n)}{\Delta t} \approx \frac{y(t_{n+1}) - y(t_n)}{\Delta t} \approx f(t_n, y(t_n))$$

possiamo calcolare $y(t_{n+1})$ conoscendo $y(t_n)$

$$ y(t_{n+1}) = y(t_n) + \Delta t \cdot f(t_n, y(t_n))$$.

Partendo da $y(t_0) = y_0$ possiamo calcolare iterativamente tutti i valori di $y(t_n)$ necessari.

Ad esempio, nel caso della legge di Malthus, $f(t_n, y(t_n)) = r\cdot y(t_n)$ e 

$$\begin{cases}
  y(t_0) = y_0 \\
  y(t_1) = y(t_0) + \Delta t\cdot ry(t_0)\\
  y(t_2) = y(t_1) + \Delta t\cdot ry(t_1)\\
  y(t_3) = y(t_2) + \Delta t\cdot ry(t_2)\\
  ...\\
  y(t_N) = y(t_{N-1}) + \Delta t\cdot ry(t_{N-1})\\
\end{cases}$$

Il seguente codice implementa il metodo di Eulero per calcolare in modo numerico e approssimato la soluzione alla legge di Malthus. Il codice inoltre confronta la soluzione approssimata con quella esatta.

```python
from matplotlib import pyplot as plt
import numpy as np
import math


delta_t = 0.5

y_0 = 100
r = 0.1

t_eulero = []
y_eulero = []

t_0 = 0
t_max = 20

t = t_0
y = y_0

while t < t_max:
    # memorizzo il valore di y e t per il grafico
    y_eulero.append(y)
    t_eulero.append(t)
    
    # variazione approssimata di y
    y = y + r*y*delta_t
    t = t + delta_t


y_esatta = [y_0*np.e**(r*t) for t in t_eulero]


# disegna le linee dei tre grafici
plt.plot(t_eulero, y_eulero, label="Eulero", linestyle="-", marker=".")
plt.plot(t_eulero, y_esatta, label="Esatta")

# etichette agli assi
plt.xlabel("tempo (anni)")
plt.ylabel("popolazione")

# titolo e legenda
plt.title("Soluzione approssimata")
plt.legend()

# attiva una griglia
plt.grid()

plt.show()
```

## Modello II: il prelievo costante nel tempo

Aggiungiamo una complicazione al nostro modello: supponiamo che in una unit di tempo vengano rimossi $B$ esemplari dalla popolazione. L'equazione per la variazione della popolazione diventa

$$\frac{\Delta y}{\Delta t} \approx r\cdot y(t) - B$$

Questo termine complica il modello, ma lo rende pi aderente alla realt. Supponiamo, ad esempio, di voler modellizzare la popolazione di delfini delle isole Eolie, come nell'esercizio precedente. Anche in questo caso esiste una soluzione esatta, ma  molto pi complicata:

$$\begin{cases}
  \frac{\Delta y}{\Delta t} \approx r\cdot y(t) - B\\
  y(0) = y_0
\end{cases}$$

e

$$y(t) = \left( y_0 - \frac{B}{r} \right) e^{rt}+\frac{B}{r}$$

## Modello III: la competizione per le risorse ambientali
Se $r>0$ legge di Malthus ha come soluzione una funzione esponenziale crescente. Dopo un certo periodo di tempo, la popolazione assume valori molto elevati che non sono sostenibili per qualsiasi ambiente. Dobbiamo introdurre un nuovo termine che tenga conto delle risorse limitate dell'ambiente (come, per esempio, la disponibilit di cibo) e riduca il tasso di crescita netto quando la popolazione  troppo elevata.

Introduciamo un nuovo parametro $K > 0$: *la capacit portante dell'ambiente*, cio la massima popolazione che l'ambiente riesce a sostenere. Sostituiamo il tasso di crescita netto con il termine:

$$r \rightarrow r\left( 1-\frac{y(t)}{K} \right)$$

in questo modo $r$ diventa negativo quando $y(t) > K$ e la popolazione decresce di conseguenza. 

Le equazioni del problema di Cauchy diventano

$$\begin{cases}
  \frac{\Delta y}{\Delta t} \approx r\left( 1-\frac{y(t)}{K} \right)\cdot y(t) = ry(t) - \frac{r}{K}y^2(t)\\
  y(0) = y_0
\end{cases}$$

Questa equazione contiene un termine non lineare e non ha una soluzione algebrica esatta. L'unica possibilit per calcolare l'evoluzione della popolazione  utilizzare un metodo di approssimazione numerica come il metodo di Eulero.

# Modello IV: dinamica di due popolazioni
Consideriamo un sistema che contenga due popolazioni, una preda con popolazione $p(t)$ e un predatore con popolazione $s(t)$. Ad esempio, le prede possono essere le popolazioni di naselli e i predatori gli squali. Un modello di questo tipo pu essere utilizzato per modellizzare le osservazioni di Umberto D'Ancona del 1925 citate nell'introduzione.

Possiamo modellizzare l'evoluzione di un sistema di questo tipo con questo sistema di equazioni:

$$\begin{cases}
  \frac{\Delta p}{\Delta t} \approx (r_p - \sigma_p p(t) - \alpha s(t))p(t)\\
  \frac{\Delta s}{\Delta t} \approx (r_s - \sigma_s s(t) - \mu \alpha p(t))s(t)\\
  p(0) = p_0
  s(0) = s_0
\end{cases}$$

Dove i coefficienti hanno il significato:
* $r_s$ e $r_p$ sono i tassi di crescita delle due popolazioni;
* $-\alpha s(t)$ descrive il prelievo dei predatori alla popolazione di prede. $\alpha$  il *tasso pro-capite di predazione*;
* $\sigma_p = r_p/K_p$ e $sigma_s = r_s/K_s$ sono coefficienti legati alla capacit portante dell'ambiente;
* $\mu$  il tasso di conversione di prede in predatori e modellizza il fatto che un'abbondanza di prede favorisce la crescita della popolazione di predatori.
