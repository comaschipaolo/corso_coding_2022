# Il metodo di bisezione
## Risolvere equazioni in modo approssimato

Nel notebook 3, quando abbiamo introdotto il ciclo `while`, abbiamo provato a indovinare un numero segreto andando per tentativi. La strategia ottimale per indovinare il numero con il numero minimo di tentativi prevede di sfruttare ogni tentativo errato per ridurre della met i possibili valori del numero segreto. Ad esempio, se dobbiamo indovinare un numero da 1 a 100, come primo tentativo  conveniente il numero 50. In questo modo ci sono tre possibilit:

*	50  il numero segreto;
*	il numero segreto  minore di 50. Allora abbiamo ristretto le possibilit a tutti i numeri compresi tra 1 e 49;
*	il numero segreto  maggiore di 50. In questo caso le possibilit sono i numeri tra 51 e 100.

Nel caso peggiore, se prima avevamo 100 possibilit, dopo il tentativo ne abbiamo 49. Questa strategia si pu applicare iterativamente, arrivando a trovare il numero segreto in, al massimo, $\log_2(100)\approx7$ tentativi (questo perch $2^7=128$).

Questa stessa idea  utilizzata per risolvere in modo approssimato qualsiasi equazione. Consideriamo un software di calcolo grafico, come ad esempio *Geogebra*: ogni volta che tracciamo il grafico di una funzione, per quanto complessa, il programma trova sempre i valori di intersezione con lasse delle $x$, in modo approssimato ma con grande precisione. Ad esempio, per la funzione trascendente $y =  \frac{\sin(x)}{x}$:




Per calcolare i punti di intersezione di un grafico con lasse delle x, dato che y=0 sullasse delle x, dobbiamo risolvere lequazione: 
$\frac{\sin(x)}{x} = 0$.
Questa  un'equazione *trascendente*, cio unequazione che non  possibile esprimere mediante polinomi e radicali. Dal punto di vista matematico , quindi, un oggetto molto complesso. Ovviamente, noi esseri umani siamo in grado di riconoscere che lequazione  soddisfatta da tutti i multipli interi di , ma un programma come Geogebra non dispone della nostra intelligenza umana, che  in grado di riconoscere schemi con un colpo docchio. Per questo motivo Geogebra ricorre a metodi generali in grado di approssimare le radici indipendentemente dalla funzione rappresentata (nella figura, infatti, Geogebra non riconosce che la radice evidenziata  , tuttavia ne mostra il suo valore con estrema precisione).

Consideriamo la funzione $f(x) = \frac{\sin(x)}{x}$ in figura nellintervallo [2, 4]:
 


I valori di $f(x)$ negli estremi dellintervallo sono:

*	$f(2)\approx 0,45 >0$;
*	$f(4)\approx-0,19 < 0$.

$f(x)$ passa da essere positiva a essere negativa. Se la funzione  continua, cio il suo grafico pu essere tracciato con una penna senza mai staccare la punta dal foglio, il grafico deve intersecare lasse delle $x$ almeno una volta. Possiamo quindi affermare che lequazione $f(x)=0$ ha almeno una soluzione nellintervallo [2, 4]. Possiamo dire che una radice dellequazione  

$$\bar{x} = 3 \pm 1$$

dove 3  il punto medio dellintervallo.
Il nostro obiettivo  restringere lestensione di questo intervallo, in modo da conoscere il valore di questa soluzione con lapprossimazione migliore possibile.

Consideriamo $x = 3$, il punto medio dellintervallo:



$f(3) \approx 0,05 >0$, quindi il valore di $f(x)$ nel punto medio ha lo stesso segno del valore nellestremo sinistro e segno opposto a quello dellestremo destro. Ragionando come prima, possiamo affermare che una radice dellequazione esiste sicuramente nellintervallo [3, 4], quindi

$$\bar{x} = 3,5 \pm 0,5$$

abbiamo dimezzato lestensione dellintervallo e la nostra incertezza sulla radice cercata.
